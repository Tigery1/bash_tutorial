```python
import matplotlib.pyplot as plt

# Data

zero_emissions_tng = 30 
total_tng = 100


sizes = [zero_emissions_tng, total_tng - zero_emissions_tng]


colors = [(0, 1, 0, 0.6), (1, 0, 0, 0.6)]  # Green and red with alpha value of 0.6

# Calculate the percentage and round to one decimal place
percentage = round((zero_emissions_tng / total_tng) * 100, 1)

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, colors=colors, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 16})
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title(f"{percentage}% of Alberta's Electricity is Currently from\n Zero Emissions Sources", fontsize=24)
plt.show()
```
